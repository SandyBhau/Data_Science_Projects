```python
import requests
from bs4 import BeautifulSoup
import urllib3

# Disable SSL certificate verification (not recommended for production)
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

def fetch_google_search_results(query):
    # Set user-agent header to mimic a browser request
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }

    # Google search URL
    url = f"https://www.google.com/search?q={query}"

    # Send GET request to Google search
    response = requests.get(url, headers=headers, verify=False)  # Disable SSL verification

    # Parse the HTML response
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all search result links
    search_results = soup.find_all("div", class_="tF2Cxc")

    # Extract top 5 links with name and description
    results = []
    for result in search_results[:5]:
        link = result.find("a")["href"]
        name = result.find("h3").get_text()
        description = result.find("span", class_="aCOpRe").get_text()
        results.append({"name": name, "link": link, "description": description})

    return results

# Example usage
query = "data science"
results = fetch_google_search_results(query)
for idx, result in enumerate(results, start=1):
    print(f"{idx}. {result['name']}")
    print(f"   Link: {result['link']}")
    print(f"   Description: {result['description']}")
    print()
```

```python
import requests
from bs4 import BeautifulSoup
import webbrowser

# Define the URL of the webpage you want to scrape
url = 'https://google.com'

# Define the headers to mimic Internet Explorer
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; Trident/7.0; rv:11.0) like Gecko'
}

# Send a GET request to the website with the defined headers
response = requests.get(url, headers=headers)
```

```python


# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the webpage using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Perform web scraping operations with Beautiful Soup
    # For example, extract text from HTML elements
    title = soup.title.get_text()
    print('Title:', title)

    # Open the webpage in Internet Explorer
    with open('temp.html', 'w', encoding='utf-8') as f:
        f.write(response.text)
    webbrowser.get('C:\\Program Files (x86)\\Internet Explorer\\iexplore.exe').open('temp.html')

else:
    print('Failed to fetch webpage:', response.status_code)
```
